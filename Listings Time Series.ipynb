```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import seaborn as sns
%matplotlib inline
```

```python
skipList = {
    2015: {
        2: True,
        4: True,
        7: True
    },
    2016: {
        3: True
    },
    2019: {
        7: True,
        8: True,
        9: True,
        10: True,
        11: True,
        12: True
    },
}

listing_df = pd.DataFrame({
    "date": [],
    "year": [],
    "month": [],
    "NYC": [],
    "Manhattan": [],
    "Brooklyn": [],
    "Queens": [],
    "Staten Island": [],
    "Bronx": []
})

listCounts = []

listCounts_Man = []
listCounts_Broo = []
listCounts_Quee = []

for yr in range(2015, 2020):
    for mo in range(1,13):
        if yr == 2019 and mo > 6:
            break
        elif yr in skipList and mo in skipList[yr]:
            # do nothing
            print('skipped')
        else:
            zeroStr = '0' if mo < 10 else ''
            df_list = pd.read_csv('./data/listings/listings' + '_' + zeroStr + str(mo) + '_' + str(yr) + '.csv')
            
            
#             print(len(df_list['id'].unique()))
#             print(len(df_list[df_list.neighbourhood_group == 'Manhattan']['id'].unique()))
#             print(len(df_list[df_list.neighbourhood_group == 'Brooklyn']['id'].unique()))
#             print(len(df_list[df_list.neighbourhood_group == 'Queens']['id'].unique()))
            
            dateObj = pd.to_datetime('1/'+str(mo)+'/'+str(yr), dayfirst=True)
    
            df_temp = pd.DataFrame({
                "date": [dateObj],
                "year": [dateObj.year],
                "month": [dateObj.month],
                "NYC": [len(df_list['id'].unique())],
                "Manhattan": [len(df_list[df_list.neighbourhood_group == 'Manhattan']['id'].unique())],
                "Brooklyn": [len(df_list[df_list.neighbourhood_group == 'Brooklyn']['id'].unique())],
                "Queens": [len(df_list[df_list.neighbourhood_group == 'Queens']['id'].unique())],
                "Staten Island": [len(df_list[df_list.neighbourhood_group == 'Staten Island']['id'].unique())],
                "Bronx": [len(df_list[df_list.neighbourhood_group == 'Bronx']['id'].unique())]
            })
            
            listing_df = listing_df.append(df_temp)
            

#             listCounts.append()

#             listCounts_Man.append(len(df_list[df_list.neighbourhood_group == 'Manhattan']['id'].unique()))
#             listCounts_Broo.append(len(df_list[df_list.neighbourhood_group == 'Brooklyn']['id'].unique()))
#             listCounts_Quee.append(len(df_list[df_list.neighbourhood_group == 'Queens']['id'].unique()))

listing_df = listing_df.set_index("date")

print(listing_df.head())
print(listing_df.shape)
```

```python
sns.set(rc={'figure.figsize': (11,4)})
```
